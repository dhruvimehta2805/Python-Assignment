```python
import pandas as pd

file_path = "Traffic_Crashes_-_Crashes.csv"
df = pd.read_csv(file_path)
df
```

```python
df.head()
```

```python
df.info()
```

```python
# Convert CRASH_DATE to datetime
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'], errors='coerce')
df
```

```python
# Handling missing values (Dropping rows where LATITUDE & LONGITUDE are missing)
df_cleaned = df.dropna(subset=['LATITUDE', 'LONGITUDE'])
df
```

```python
# Removing potential incorrect latitude & longitude values
df = df_cleaned.dropna()
df = df_cleaned[(df_cleaned['LATITUDE'] != 0) & (df_cleaned['LONGITUDE'] != 0)]
df
```

```python
# Handling outliers (Removing rows with unusually high injury counts)
df_cleaned = df_cleaned[df_cleaned['INJURIES_TOTAL'] <= 10]  # Keeping values below 10 for reasonability
df
```

```python
# Save the cleaned dataset
df_cleaned.to_csv("cleaned_traffic_dataset.csv", index=False)
df
print("Data Cleaning Completed. Cleaned file saved as 'cleaned_traffic_dataset.csv'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned traffic data
file_path = 'cleaned_traffic_dataset.csv'
data = pd.read_csv(file_path)
```

```python
# Convert date columns to datetime format if they exist
data['CRASH_DATE'] = pd.to_datetime(data['CRASH_DATE'], errors='coerce')

# 1. Number of accidents per year
plt.figure(figsize=(8, 6))
data['CRASH_DATE'].dt.year.value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Accidents Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# 2. Number of accidents per month (line plot)
plt.figure(figsize=(8, 6))
data['CRASH_DATE'].dt.month.value_counts().sort_index().plot(kind='line', color='salmon', marker='o')
plt.title('Number of Accidents Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# 3. Distribution of accidents by day of week (pie chart)
plt.figure(figsize=(8, 6))
data['CRASH_DATE'].dt.dayofweek.value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'yellowgreen', 'darkgreen'])
plt.title('Accidents by Day of Week (0=Monday, 6=Sunday)')
plt.ylabel('')
plt.show()
```

```python
# 4. Histogram: Distribution of crash hours
plt.figure(figsize=(8, 6))
plt.hist(df_cleaned['CRASH_HOUR'].dropna(), bins=24, color='skyblue', edgecolor='black')
plt.title('Distribution of Crashes by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Crashes')
plt.show()
```

```python
# 5. Weather conditions vs number of collisions
plt.figure(figsize=(8, 6))
df_cleaned['WEATHER_CONDITION'].value_counts().plot(kind='bar', color='cornflowerblue')
plt.title('Collisions by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Collisions')
plt.show()
```

```python
# 6.# Stacked bar chart: Injury types distribution
plt.figure(figsize=(8, 6))
df_cleaned[['INJURIES_FATAL', 'INJURIES_INCAPACITATING', 'INJURIES_NON_INCAPACITATING']].sum().plot(kind='bar', stacked=True, color=['red', 'orange', 'yellow'])
plt.title('Injury Types Distribution')
plt.xlabel('Injury Type')
plt.ylabel('Total Count')
plt.show()
```

```python
# 7. Count plot: Types of crashes
plt.figure(figsize=(8, 6))
df_cleaned['CRASH_TYPE'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Types of Crashes')
plt.xlabel('Crash Type')
plt.ylabel('Number of Crashes')
plt.show()
```

```python
# 8. Heatmap-like plot: Crashes by hour and day
crash_hour_day = df_cleaned.groupby(['CRASH_HOUR', 'CRASH_DAY_OF_WEEK']).size().unstack(fill_value=0)
plt.imshow(crash_hour_day, aspect='auto')
plt.title('Crashes by Hour and Day')
plt.colorbar()
plt.show()
```

```python
# 9. Horizontal bar graph: Collisions by road surface condition
plt.figure(figsize=(8, 6))
df_cleaned['ROADWAY_SURFACE_COND'].value_counts().plot(kind='barh', color='tomato')
plt.title('Collisions by Road Surface Condition')
plt.xlabel('Number of Collisions')
plt.ylabel('Road Surface Condition')
plt.show()
```

```python
# 10. Line plot: Injuries and fatalities over time
plt.figure(figsize=(8, 6))
df_cleaned.groupby(df_cleaned['CRASH_DATE'].dt.year)['INJURIES_TOTAL'].sum().plot(kind='line', color='teal', marker='o')
plt.title('Total Injuries Over Time')
plt.xlabel('Year')
plt.ylabel('Total Injuries')
plt.show()
```

```python
# 11. Hit and run cases (pie chart)
plt.figure(figsize=(6, 6))
df_cleaned['HIT_AND_RUN_I'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['darkorange', 'lightcoral'])
plt.title('Hit and Run Cases')
plt.ylabel('')
plt.show()
```

```python
# 12. Histogram: Vehicles in crashes
plt.hist(df_cleaned['NUM_UNITS'], bins=8)
plt.title('Vehicles Involved in Crashes')
plt.show()
```

```python
# 13. Scatter plot: Speed limit vs injuries
plt.scatter(df_cleaned['POSTED_SPEED_LIMIT'], df_cleaned['INJURIES_TOTAL'], alpha=0.5)
plt.title('Speed Limit vs Total Injuries')
plt.show()
```

```python
# 14. Weekend vs weekday accidents (bar chart)
plt.figure(figsize=(8, 6))
data['CRASH_DATE'].dt.weekday.apply(lambda x: 'Weekend' if x >= 5 else 'Weekday').value_counts().plot(kind='bar', color='seagreen')
plt.title('Weekend vs Weekday Accidents')
plt.xlabel('Day Type')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# 15. Horizontal bar: Top 10 crash causes
df_cleaned['PRIM_CONTRIBUTORY_CAUSE'].value_counts().head(10).plot(kind='barh')
plt.title('Top 10 Crash Causes')
plt.show()
```

```python
# 16. Bar chart: Severity of accidents
plt.figure(figsize=(8, 6))
df_cleaned[['INJURIES_FATAL', 'INJURIES_TOTAL']].sum().plot(kind='bar', color=['red', 'blue'])
plt.title('Severity of Accidents')
plt.xlabel('Injury Type')
plt.ylabel('Count')
plt.show()
```

```python
# 17. Crash locations
plt.scatter(df['LATITUDE'], df['LONGITUDE'], s=df['INJURIES_TOTAL']*10, alpha=0.5)
plt.title('Bubble Chart of Crash Locations')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

```python
# 18. Scatter plot: Time of day impact on injuries
plt.figure(figsize=(8, 6))
plt.scatter(df_cleaned['CRASH_DATE'].dt.hour, df_cleaned['INJURIES_TOTAL'], alpha=0.5, color='crimson')
plt.title('Impact of Time of Day on Injuries')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Injuries')
plt.show()
```

```python
#19. Line plot: Traffic signal types
plt.figure(figsize=(8, 6))
plt.plot(df_cleaned['TRAFFIC_CONTROL_DEVICE'].value_counts(), marker='o')
plt.title('Traffic Signal Types')
plt.xlabel('Traffic Control Device')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right') #horizontal alignment
plt.show()
```

```python
#20. Top 10 Streets with Most Crashes (Bar Chart)
plt.figure(figsize=(8, 6))
top_streets = df_cleaned['STREET_NAME'].value_counts().head(10)
top_streets.plot(kind='bar', color='skyblue')
plt.title('Top 10 Streets with Most Crashes')
plt.xlabel('Street Name')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45)
plt.show()
```
